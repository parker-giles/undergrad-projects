```python
# CELL 1: Importing libaries and audio file, normalization, and waveform plotting
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

sample_rate, audio = wavfile.read("guitar_part.wav") # This pulls our mono .wav 

audio = audio.astype(np.float32) # This normalizes our audio  
audio /= np.max(np.abs(audio))

t = np.arange(len(audio)) / sample_rate # This is time for our plot in seconds

plt.figure(figsize=(12, 4))
plt.plot(t, audio)
plt.title("Waveform")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
# CELL 2: FFT of audio and frequency, magnitudes, and frequency spectrum plotting
fft_audio = np.fft.fft(audio) # This performs the FFT on the audio

mags = np.abs(fft_audio) # This gives us the magnitudes which we can work with

freqs = np.fft.fftfreq(len(audio), 1/sample_rate) 

plt.plot(freqs, mags)
plt.title("Frequency Spectrum")
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
```

```python
# CELL 3: Compression stage for 5%: keeping the top 5% most important frequencies and zeroing out the rest
threshold_ratio_a = 0.05 

k = int(len(mags) * threshold_ratio_a)
threshold_a = np.partition(mags, -k)[-k]

fft_audio_new_a = fft_audio.copy()
fft_audio_new_a[mags < threshold_a] = 0
```

```python
# CELL 4: Compression stage for 25%: keeping the top 25% most important frequencies and zeroing out the rest 
threshold_ratio_b = 0.25 

k = int(len(mags) * threshold_ratio_b)
threshold_b = np.partition(mags, -k)[-k]

fft_audio_new_b = fft_audio.copy()
fft_audio_new_b[mags < threshold_b] = 0
```

```python
# CELL 5: Reconstructs the top 5% verison, renormalizes it, converts back to mono audio, exports it, and plots the compressed waveform
reconstructed_audio_a = np.fft.ifft(fft_audio_new_a).real # This performs the IFFT

reconstructed_audio_a /= np.max(np.abs(reconstructed_audio_a)) # Renormalize

p5_guitar_part = (reconstructed_audio_a * 32767).astype(np.int16) # Brings us back to int16

wavfile.write("5p_guitar_part.wav", sample_rate, p5_guitar_part)

tt_a = np.arange(len(reconstructed_audio_a)) / sample_rate 

plt.figure(figsize=(12, 4))
plt.plot(tt_a, reconstructed_audio_a)
plt.title("Waveform of Top 5%")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
# CELL 6: Reconstructs the top 25% verison, renormalizes it, converts back to mono audio, exports it, and plots the compressed waveform
reconstructed_audio_b = np.fft.ifft(fft_audio_new_b).real # This performs the IFFT

reconstructed_audio_b /= np.max(np.abs(reconstructed_audio_b)) # Renormalize

p25_guitar_part = (reconstructed_audio_b * 32767).astype(np.int16) # Brings us back to int16

wavfile.write("25p_guitar_part.wav", sample_rate, p25_guitar_part)

tt_b = np.arange(len(reconstructed_audio_b)) / sample_rate 

plt.figure(figsize=(12, 4))
plt.plot(tt_b, reconstructed_audio_b)
plt.title("Waveform of Top 25%")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```
